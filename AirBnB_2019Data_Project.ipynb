## Data Transformation

```python
sns.histplot('reviews_per_month',x=df['reviews_per_month'], kde=True)
plt.show()
```

```python
# iqr method
q1 = df.reviews_per_month.quantile(0.25)
q3 = df.reviews_per_month.quantile(0.75)

iqr = q3-q1
iqr
```

```python
lower = q1 - 1.5*iqr
upper = q3 + 1.5*iqr

# outliers
df[(df.reviews_per_month<lower) | (df.reviews_per_month>upper)][['reviews_per_month']].head()
```

```python
# Calculate the IQR
Q1 = df['reviews_per_month'].quantile(0.25)
Q3 = df['reviews_per_month'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame to remove outliers
df_iqr_transformed = df[(df['reviews_per_month'] >= lower_bound) & (df['reviews_per_month'] <= upper_bound)]

# Create a boxplot to visualize the IQR
plt.figure(figsize=(8, 5), dpi=100)
sns.boxplot(data=df, x='reviews_per_month')
plt.title('Distribution of Reviews per Month (with Outliers)')
plt.xlabel('Reviews per Month')
plt.show()

# Create a histogram of the transformed data
plt.figure(figsize=(8, 5), dpi=100)
sns.histplot(data=df_iqr_transformed, x='reviews_per_month', kde=True)
plt.title('Distribution of Reviews per Month (Without Outliers)')
plt.xlabel('Reviews per Month')
plt.ylabel('Frequency')
plt.show()
```

```python
# Group the data by 'neighbourhood_group' and count the number of listings in each group
grouped = df['neighbourhood_group'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 6), dpi=100)
explode = (0, 0, 0.2, 0.2, 0.2)  # Explode the 3rd, 4th, and 5th slices
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0']

grouped.plot(kind='pie', labels=None, autopct='%1.1f%%', startangle=90, pctdistance=0.85, explode=explode, colors=colors)

# Add labels
plt.title('Distribution of Listings in Neighbourhood Groups')
plt.legend(labels=grouped.index, bbox_to_anchor=(1, 1), loc='upper left')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    
plt.show()
```

```python
# Get the top 10 neighbourhoods with the most listings
top_10_neighbourhoods = df['neighbourhood'].value_counts().head(10)

# Create a bar plot using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_neighbourhoods.index, y=top_10_neighbourhoods.values, palette='viridis')

# Add labels
plt.title('Top 10 Neighbourhoods with the Most Listings')
plt.xlabel('Neighbourhood')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

plt.show()
```

```python
# Count the number of listings for each room type
room_type_counts = df['room_type'].value_counts()

# Define colors for each room type
colors = ['gold', 'lightcoral', 'lightskyblue']

# Create a pie chart with custom colors
plt.figure(figsize=(8, 6))
plt.pie(room_type_counts, labels=room_type_counts.index, autopct='%.2f %%', pctdistance=0.8, colors=colors)

# Add labels
plt.title('Distribution of Room Types')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
## Group the data by 'host_id' and calculate the average of price
df_listings_by_Hosts = df.groupby(['host_id', 
                                   'neighbourhood_group', 
                                   'calculated_host_listings_count'])['price'] \
                    .mean().reset_index(name='Average_price')

df_listings_by_Hosts.head()
```

```python
# Define a custom color palette
custom_palette = {
    'Manhattan': 'red',
    'Brooklyn': 'blue',
    'Queens': 'green',
    'Bronx': 'purple',
    'Staten Island': 'orange'
}

# Assuming df_listings_by_Hosts contains your data

plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(x='calculated_host_listings_count', 
                          y='Average_price',
                          data=df_listings_by_Hosts,
                          hue='neighbourhood_group',
                          palette=custom_palette)  # Use the custom color palette

plt.title("Calculated_host_listing_count Vs. Average price")
plt.show()
```

```python
# Define custom colors for each 'neighbourhood_group'
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'lightpink']

# Group the data by 'neighbourhood_group' and calculate the total price
df_PriceByNeighbourhoodGroup = df.groupby('neighbourhood_group')['price'].sum().reset_index()

# Create a pie chart with custom colors
plt.figure(figsize=(8, 8))
plt.pie(df_PriceByNeighbourhoodGroup['price'], labels=df_PriceByNeighbourhoodGroup['neighbourhood_group'],
        autopct='%.2f%%', pctdistance=0.85, colors=colors)

# Add a title
plt.title('Total Price Distribution in NYC', loc="left")

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
corr = df_iqr_transformed[['price',  
                           'minimum_nights', 
                           'number_of_reviews', 
                           'reviews_per_month', 
                           'calculated_host_listings_count', 
                           'availability_365']].corr()

corr
```

```python
# You can customize the color palette by specifying 'cmap'

# Define a custom color palette (e.g., 'coolwarm')
custom_cmap = sns.color_palette('coolwarm')

# Create a heatmap with the custom color palette
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap=custom_cmap)

# Add a title
plt.title('Correlation Heatmap with Custom Color Palette')

plt.show()
```

```python
# Create a list of unique neighborhood groups for colors
neighborhood_groups = df['neighbourhood_group'].unique()
colors = plt.cm.jet(np.linspace(0, 1, len(neighborhood_groups)))

# Create a scatter plot with different colors and sizes based on 'price'
plt.figure(figsize=(10, 10), dpi=100)
for group, color in zip(neighborhood_groups, colors):
    subset = df[df['neighbourhood_group'] == group]
    plt.scatter(subset['longitude'], subset['latitude'], label=group, s=subset['price'] * 0.1, color=color, alpha=0.6)

# Add labels
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.title("Location of listings")

# Create a legend
plt.legend()

# Show plot
plt.show()
```
